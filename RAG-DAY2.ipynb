```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings   
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.llms import Ollama
from langchain.chains import RetrievalQA

#Step 1: Loading the data
loader= TextLoader("my_docs.txt")
docs=loader.load()

#Step2: Split the document into multiplechunk
obj=CharacterTextSplitter(chunk_size=5,chunk_overlap=2)
obj.split_documents(docs)

# Step 3: Embedding
embedding_obj=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Step 4: Store in database
vectorstore= FAISS.from_documents(docs,embedding_obj)

# Step 5: Retriever
retriever_obj= vectorstore.as_retriever()

# Step 6: use LLM model
llm_obj = Ollama(model="gemma2:2b")

# Step 7: Build retriver QA Chain
RetrievalQA.from_chain_type(llm=llm_obj, retriever=retriever_obj, return_source_documents=True)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm_obj, retriever=retriever_obj, return_source_documents=True)
response= qa_chain('What is LangSmith?')
print(response)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings   
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.llms import Ollama
from langchain.chains import RetrievalQA

#Step 1: Loading the data
loader= TextLoader("my_docs.txt")
docs=loader.load()

#Step2: Split the document into multiplechunk
obj=CharacterTextSplitter(chunk_size=5,chunk_overlap=2)
obj.split_documents(docs)

# Step 3: Embedding
embedding_obj=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Step 4: Store in database
vectorstore= FAISS.from_documents(docs,embedding_obj)

# Step 5: Retriever
retriever_obj= vectorstore.as_retriever()

# Step 6: use LLM model
llm_obj = Ollama(model="gemma2:2b")

# Step 7: Build retriver QA Chain
RetrievalQA.from_chain_type(llm=llm_obj, retriever=retriever_obj, return_source_documents=False)
```
